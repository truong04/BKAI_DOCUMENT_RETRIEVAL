```python
model = SentenceTransformer('/kaggle/working/checkpoint-24000')
```

```python
batch_size = 1
all_embeddings = []
all_cids = []
final_path = "embeddings"
print("start")
for i in tqdm(range(0, len(data), batch_size), desc="Processing Batches"):
    sentences = data.iloc[i:i + batch_size]['tokens'].tolist()  # Lấy batch hiện tại
    cids = data.iloc[i:i + batch_size]['cid'].tolist() 
    with torch.no_grad():
         embeddings = model.encode(sentences, show_progress_bar = False)
        
    if embeddings is not None:
        all_embeddings.append(embeddings)
        all_cids.append(cids)
        
all_embeddings = np.vstack(all_embeddings)  
all_cids = np.array(all_cids)               

np.save(final_path + '_embeddings.npy', all_embeddings)  # Lưu embeddings
np.save(final_path + '_cids.npy', all_cids)          
print("Encoding and saving completed!")
```
